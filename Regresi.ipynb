```python
# import library yang dibutuhkan
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import sklearn
```

```python
# memanggil dataset
dataset = pd.read_csv('/content/invoice_retail_dataset.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, 1].values
```

```python
x = dataset[['UnitPrice']].values
y = dataset[['CustomerID']].values
```

```python
dataset.keys()
```

```python
dataset.shape
```

```python
# Menampilkan isi sebagian dataset
dataku = pd.DataFrame(dataset)
dataku.head()
```

```python
# Split Dataset menjadi Training Set dan Testing Set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=80)
```

```python
# Melakukan Fitting Simple Linear Regression pada Training Set
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(x_train, y_train)
```

```python
# Memprediksi hasil Test-Set
y_pred = regressor.predict(x_test)
```

```python
# Visualisasi Data
plt.scatter(dataku.UnitPrice, dataku.CustomerID)
plt.xlabel("Unit Price")
plt.ylabel("Customer ID")
plt.title("Unit Price vs Customer ID")
plt.show
```

```python
# Visualisasi Hasil Prediksi Pada Training-Set
# Ukuran Plot
plt.figure(figsize=(10,8))

# Biru adalah data observasi
plt.scatter(x_train, y_train, color = 'blue')

# Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_train, regressor.predict(x_train), color = 'red')

# Memberi Judul dan Label
plt.title('Harga Satuan Barang terhadap ID Pelanggan (Training Set)')
plt.xlabel('Unit Price')
plt.ylabel('Customer ID')

# Menampilkan plot
plt.show()
```

```python
# Visualisasi Hasil Prediksi Pada Test-Set
# Ukuran Plot
plt.figure(figsize=(10,8))

# Biru adalah data observasi
# plt.scatter(x_test, y_test, color = 'blue')
plt.scatter(np.arange(len(x_test)), y_test, color = 'blue')

# Garis merah adalah hasil prediksi dari machine learning
plt.plot(x_test, regressor.predict(x_test), color = 'red')

# Memberi Judul dan Label
plt.title('Harga Satuan Barang terhadap ID Pelanggan (Training Set)')
plt.xlabel('Unit Price')
plt.ylabel('Customer ID')

# Menampilkan plot
plt.show()
```
